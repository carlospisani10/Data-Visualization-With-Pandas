```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import sem, ttest_ind
```

```python
general_heights = pd.read_csv("../Resources/general_heights.csv")
general_heights
```

```python
general_heights = general_heights.iloc[:,0]
general_heights
```

```python
wnba_data = pd.read_csv("../Resources/wba_data.csv")
wnba_heights = wnba_data.iloc[:, -1]
wnba_heights
```

```python
(t_stat, p) = ttest_ind(general_heights, wnba_heights, equal_var=False)
```

```python
print("The mean height of WNBA players is " + str(wnba_heights.mean()))
print("The mean height of women sampled is " + str(general_heights.mean()))
```

```python
if p < 0.05:
    print("The difference in sample means is significant.")
else:
    print("The difference in sample means is not significant.")
```

```python
tick_labels = ["General Public", "WBA Players"]

means = [general_heights.mean(), wnba_heights.mean()]
x_axis = np.arange(0, len(means), 1)
means
```

```python
sem = [sem(general_heights), sem(wnba_heights)]
sem
```

```python
plt.errorbar(x_axis, means, sem, fmt="o")
plt.title("Mean Height of Women in General Population and WBA Players")

plt.xlim(-0.5, 1.5)
plt.ylim(64, 73)

plt.xticks(x_axis, tick_labels)

plt.ylabel("Height (Inches)")

plt.show()
```
